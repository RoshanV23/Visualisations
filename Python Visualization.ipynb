```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
countries_gdp=pd.DataFrame(countries_gdp)
countries_gdp
```

```python
plt. figure(figsize=(8,6))

plt.bar(countries_gdp['Name'],countries_gdp['GDP'],color='orange')
plt.xlabel('Name of the Country')
plt.ylabel('GDP of the Country')
plt.title('GDP of 4 Countries')
```

```python
tips=pd.read_csv('tips.csv')
tips
```

```python
plt.figure(figsize=(8,6))

plt.scatter(tips['total_bill'], tips['tip'], color = 'blue')

plt.xlabel('Bill Amount in $')
plt.ylabel('Tip Amount in $')
plt.title('Bills and Tips')
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x = 'total_bill', y= 'tip' , data=tips , hue = 'sex')
```
